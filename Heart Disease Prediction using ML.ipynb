```python
#Importing the Dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the csv data as a Pandas Dataframe

heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
#print first five rows of the dataset
heart_data.head()
```

```python
#print last five rows of the dataset
heart_data.tail()
```

```python
#checking the number of rows and columns in the dataset
heart_data.shape
```

```python
#getting some info about the data
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#Statistical measures of the data

heart_data.describe()
```

```python
# Checking the distribution of target variable

heart_data['target'].value_counts()
```

1--> Defective Heart
0--> Healthy Heart

```python
#Splitting hte features and Target

X = heart_data.drop(columns='target',axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training data and test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING

```python
#Logistic Regression Model

model = LogisticRegression()
```

```python
#training the logistic regression model with training data

model.fit(X_train,Y_train)
```

Model Evaluation

```python
#Accuracy Score
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print(f"Accuracy on Training data is: {training_data_accuracy*100} %")
```

```python
#Accuracy Score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print(f"Accuracy on Test data is: {test_data_accuracy*100} %")
```

Predictive System

```python
input_data = (53,1,2,130,197,1,0,152,0,1.2,0,0,2)

#Change input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshaping the numpy array as we are prediciting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)

if(prediction[0]==0):
    print("The person doesn't have a Heart Disease")
else:
    print("The person has Heart Disease")    
```
